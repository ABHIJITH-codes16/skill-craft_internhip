```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('matches.csv')
season_counts = df['season'].value_counts().sort_index()

plt.bar(season_counts.index, season_counts.values)
plt.xlabel('Season')
plt.ylabel('Number of Matches')
plt.title('Number of IPL Matches per Season')
plt.show()

```

```python
import pandas as pd

df = pd.read_csv('deliveries.csv')

extras_by_bowler = df.groupby('bowler')['extra_runs'].sum()
highest_extras = extras_by_bowler.sort_values(ascending=False).head(1)

print("Highest extras bowled by a player:")
print(highest_extras)
```

```python
import matplotlib.pyplot as plt

top5 = extras_by_bowler.sort_values(ascending=False).head(5)


```

```python
import matplotlib.pyplot as plt

top10 = extras_by_bowler.sort_values(ascending=False).head(10)


plt.figure(figsize=(10, 6))
plt.barh(top10.index, top10.values)
plt.xlabel('Total Extra Runs')
plt.ylabel('Bowler')
plt.title('Top 10 Bowlers with Highest Extras in IPL')
plt.gca().invert_yaxis()   # Highest on top
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()

```
